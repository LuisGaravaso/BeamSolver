### Importar Módulo

```python
from vigas import *
```

### Descrever Carregamento

```python
L = 4
q0 = 500
Q = Sing(-q0, L/4, expo = 0) + Apoio(L/4) + Apoio(3*L/4) + Sing(q0, 3*L/4, expo = 0) 
```

### Equação do Carregamento

```python
Q
```

### Viga

```python
viga = Viga(Q, L = L, E = 200e9, Izz = 2004e-8)
```

### Condições de Contorno

```python
viga.add_forca(mag = 0, pos = "i")
viga.add_forca(mag = 0, pos = "f")
viga.add_momento(mag = -150, pos = "i")
viga.add_momento(mag = -150, pos = "f")
```

### Solver

```python
viga.solve()
```
